1. Write a python program which searches all the product under a particular product from www.amazon.in. The 
product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for 
guitars.

```python
driver=webdriver.Chrome()
driver.get("https://www.amazon.com")
product = input("Enter the product you want to search : ")
search=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
search.send_keys(product)
searchbtn=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
searchbtn.click()
```

2. In the above question, now scrape the following details of each product listed in first 3 pages of your search 
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then 
scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
products = []
for page in range(1, 4):
    time.sleep(2)  
    product_containers = driver.find_elements(By.XPATH, '//div[@data-component-type="s-search-result"]')

    for container in product_containers:
        try:
            nametitle = container.find_element(By.XPATH, './/span[@class="a-size-base-plus a-color-base a-text-normal"]').text
            maintext=nametitle.split(' ', 1)
            brand_name=maintext[0]
            product_title=maintext[1]
            
        except:
            brand_name = "-"
            product_title = "-"

        try:
            price = container.find_element(By.XPATH, './/span[@class="a-price-whole"]').text
            availability="Available"
        except:
            price = "-"
            availability="Not Available"

        try:
            product_url = container.find_element(By.XPATH, './/a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]').get_attribute('href')
        except:
            product_url = "-"

        try:
            return_exchange = container.find_element(By.XPATH, './/span[contains(text(), "Return")]').text
        except:
            return_exchange = "-"

        try:
            expected_delivery = container.find_element(By.XPATH, './/span[contains(text(), "Delivery")]').text
        except:
            expected_delivery = "-"

        products.append({
            'Brand Name': brand_name,
            'Name of the Product': product_title,
            'Price': price,
            'Return/Exchange': return_exchange,
            'Expected Delivery': expected_delivery,
            'Availability': availability,
            'Product URL': product_url
        })

    try:
        next_page = driver.find_element(By.XPATH, '//li[@class="a-last"]/a')
        next_page.click()
    except:
        break
driver.quit()
Data = pd.DataFrame(products)
Data.to_csv('amazon.csv')
Data
```

4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com
and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand 
Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, 
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
details is missing then replace it by “- “. Save your results in a dataframe and CSV. 

```python
driver=webdriver.Chrome()
driver.get("https://www.flipkart.com")
time.sleep(3)
search=driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input")
search.send_keys("smartphone")
search_btn=driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button")
search_btn.click()

Brand_name=[]
Smartphone_name=[]
Color=[]
RAM=[]
ROM=[]
Primary_camera=[]
Secondary_camera=[]
Display_size=[]
Battery_capacity=[]
Price=[]
Product_url=[]

title=driver.find_elements(By.CLASS_NAME,"KzDlHZ")
price=driver.find_elements(By.CLASS_NAME,"_4b5DiR")
displaysize=driver.find_elements(By.XPATH,"//ul[contains(@class,'G4BRas')]/li[2]")
batterycapacity=driver.find_elements(By.XPATH,"//ul[contains(@class,'G4BRas')]/li[4]")
producturl=driver.find_elements(By.CLASS_NAME,"CGtC98")
ramrom=driver.find_elements(By.XPATH,"//ul[contains(@class,'G4BRas')]/li[1]")
camera=driver.find_elements(By.XPATH,"//ul[contains(@class,'G4BRas')]/li[3]")

for i in price:
    Price.append(i.text)
for j in displaysize:
    Display_size.append(j.text)
for k in batterycapacity:
    Battery_capacity.append(k.text)
for n in producturl:
    Product_url.append(n.get_attribute('href'))
for m in title:
    split_text=re.split(r'[\(\)]', m.text)
    main_text=re.split(r'\s+', split_text[0], maxsplit=1)
    Brand_name.append(main_text[0])
    Smartphone_name.append(main_text[1])
    colorname = re.split(r',', split_text[1])
    Color.append(colorname[0])
for l in ramrom:
    result=re.split(r'\|', l.text)
    RAM.append(result[0])
    ROM.append(result[1])
for o in camera:
    result=re.split(r'\|', o.text)
    Primary_camera.append(result[0])
    if(len(result)==1):
        Secondary_camera.append("-")
    else:
        Secondary_camera.append(result[1])

driver.close()

data=pd.DataFrame({"Brand Name":Brand_name,"Smartphone Name":Smartphone_name,"Color":Color,"RAM":RAM,"Storage(ROM)":ROM,"Primary Camera":Primary_camera,"Secondary Camera":Secondary_camera,"Display size":Display_size,"Battery capacity":Battery_capacity,"Price":Price,"Product URL":Product_url})
data.to_csv('smartphones.csv')
data
```

5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 

```python
driver=webdriver.Chrome()
driver.get("https://www.google.com/maps/")
time.sleep(4)
search=driver.find_element(By.ID,"searchboxinput")
city=input("Enter your city : ")
search.send_keys(city)
search_btn=driver.find_element(By.ID,"searchbox-searchbutton")
search_btn.click()
time.sleep(3)
city_url = driver.current_url
split_url = re.split(r'@', city_url)
geospatial_coordinates=re.split(r',', split_url[1])
latitude=geospatial_coordinates[0]
longitude=geospatial_coordinates[1]
driver.close()
print("Geospatial Coordinates of "+city+" city is latitude: "+str(latitude)+", longitude: "+str(longitude))
```

6. Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
driver=webdriver.Chrome()
driver.get("https://www.digit.in/")
search=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div/div/div[2]/form/input[1]")
search.send_keys("Gaming Laptop")
search_btn=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div/div/div[2]/form/button")
search_btn.click()
time.sleep(2)
Name=[]
Price=[]
k=1
for n in range(1,4):
    name=driver.find_elements(By.XPATH,"//h3[contains(@class,' text-clamp text-clamp-2')]")
    price=driver.find_elements(By.XPATH,"//span[contains(@class,'amount')]")
    for i in name:
        Name.append(i.text)

    for j in price:
        text1=j.text
        result=text1.split('\n')
        Price.append(result[0])
    if(n<3):
        page=driver.find_element(By.XPATH,"//nav[contains(@class,'woocommerce-pagination')]/ul/li["+str(n+k)+"]")
        page.click()
        time.sleep(5)
        k+=1
driver.close()
data=pd.DataFrame({"Name":Name[:30],"Price":Price[:30]})
data
```

7. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: 
“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
driver=webdriver.Chrome()
driver.maximize_window()
driver.get("https://www.forbes.com")
time.sleep(2)
menu=driver.find_element(By.XPATH,"//div[contains(@class,'USVF4LaM')]")
menu.click()
time.sleep(1)
billionaires=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/header/div[1]/div[1]/div[2]/ul/li[2]/div[1]/span")
billionaires.click()
worlds_billionaires=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/header/div[1]/div[1]/div[2]/ul/li[2]/div[2]/div[3]/ul/li[1]/a")
worlds_billionaires.click()
time.sleep(3)

Rank=[]
Name=[]
Net_worth=[]
Age=[]
Citizenship=[]
Source=[]
Industry=[]


for t in range(14):
    rank=driver.find_elements(By.XPATH,"//div[contains(@class,'Table_tableRow__lF_cY')]/div[1]/div")
    name=driver.find_elements(By.XPATH,"//div[contains(@class,'Table_tableRow__lF_cY')]/div[2]")
    networth=driver.find_elements(By.XPATH,"//div[contains(@class,'Table_tableRow__lF_cY')]/div[3]/span")
    age=driver.find_elements(By.XPATH,"//div[contains(@class,'Table_tableRow__lF_cY')]/div[4]/div")
    citizenship=driver.find_elements(By.XPATH,"//div[contains(@class,'Table_tableRow__lF_cY')]/div[5]")
    source=driver.find_elements(By.XPATH,"//div[contains(@class,'Table_tableRow__lF_cY')]/div[6]")
    industry=driver.find_elements(By.XPATH,"//div[contains(@class,'Table_tableRow__lF_cY')]/div[7]")
    
    for i in rank:
        Rank.append(i.text)
    for j in name:
        Name.append(j.text)
    for k in networth:
        Net_worth.append(k.text)
    for n in age:
        Age.append(n.text)
    for m in citizenship:
        Citizenship.append(m.text)
    for l in source:
        Source.append(l.text)
    for p in industry:
        Industry.append(p.text)
    if(t<13):
        nextpage=driver.find_element(By.XPATH,"//button[contains(@aria-label,'Next')]")
        nextpage.click()
        time.sleep(1)


data=pd.DataFrame({"Rank":Rank,"Name":Name,"Net Worth":Net_worth,"Age":Age,"Citizenship":Citizenship,"Source":Source,"Industry":Industry})
data
```

9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in 
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall 
reviews, privates from price, dorms from price, facilities and property description.

```python
driver=webdriver.Chrome()
driver.get("https://www.hostelworld.com/")
time.sleep(2)
search=driver.find_element(By.XPATH,"/html/body/div[3]/div/div[2]/main/header/div/div[2]/div[1]/div[1]/div/div[1]/div[1]/div/div[2]/label/input")
search.send_keys("London")
time.sleep(2)
suggetion=driver.find_element(By.XPATH,"/html/body/div[3]/div/div[2]/main/header/div/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/ul/li[2]/button")
suggetion.click()
search_btn=driver.find_element(By.XPATH,"/html/body/div[3]/div/div[2]/main/header/div/div[2]/div[1]/div[1]/div/div[5]/button[2]")
search_btn.click()
time.sleep(10)

Hostel_name=[]
Distance_from_city_center=[]
Ratings=[]
Total_reviews=[]
Overall_reviews=[]
Privates_from_price=[]
Dorms_from_price=[]
Facilities=[]


hostelname=driver.find_elements(By.XPATH,"//div[contains(@class,'property-name')]/span")
distancefromcitycenter=driver.find_elements(By.XPATH,"//span[contains(@class,'distance-description')]")
ratings=driver.find_elements(By.XPATH,"//span[contains(@class,'number')]")
totalreviews=driver.find_elements(By.XPATH,"//span[contains(@class,'left-margin')]")
overallreviews=driver.find_elements(By.XPATH,"//span[contains(@class,'keyword')]")
privatesfromprice=driver.find_elements(By.XPATH,"//div[contains(@class,'property-accommodation-prices')]")
facilities=driver.find_elements(By.XPATH,"//div[contains(@class,'property-badges')]/span/div/div/div")

for j in hostelname:
    Hostel_name.append(j.text)
for k in distancefromcitycenter:
    Distance_from_city_center.append(k.text[2:])
for l in ratings:
    Ratings.append(l.text)
for l in totalreviews:
    ttext=l.text
    Total_reviews.append(ttext.replace("(", "").replace(")", ""))
for n in overallreviews:
    Overall_reviews.append(n.text)
for m in privatesfromprice:
    
    condition1="%"
    condition2="Privates From"
    condition3="No Privates Available"
    condition4="No Dorms Available"
    
    text1=m.text
    result=text1.split('\n')
    
    if condition1 in result[0]:
        Privates_from_price.append(result[3])
    if condition2 in result[0]:
        Privates_from_price.append(result[1])
    if condition3 in result[0]:
        Privates_from_price.append("-")
        
    if(len(result)==8):
        Dorms_from_price.append(result[7])
    if(len(result)==4):
        Dorms_from_price.append(result[3])
    if(len(result)==3):
        if condition4 in result[2]:
            Dorms_from_price.append("-")
        else:
            Dorms_from_price.append(result[2])
    if(len(result)==5):
        Dorms_from_price.append(result[4])
        
for p in facilities:
    Facilities.append(p.get_attribute('arialabel'))
     
driver.close()
data=pd.DataFrame({"Hostel name":Hostel_name,"Distance from city center":Distance_from_city_center,"Ratings":Ratings[2:],"Total reviews":Total_reviews,"Overall reviews":Overall_reviews[2:],"Privates from price":Privates_from_price,"Dorms from price":Dorms_from_price,"Facilities":Facilities[:30]})
data
```
