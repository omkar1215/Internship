**Question 1**- Write a python program to display IMDB’s Top rated 100 Indian movies’ data
https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year ofrelease) and make data frame.

```python
#Answer 1
#Before executing this cell, execute the first cell to import the required libraries.
#Answer 1
#Before executing this cell, execute the first cell to import the required libraries.
page=requests.get('https://www.imdb.com/list/ls056092300/')
print(page)
soup=BeautifulSoup(page.content)
name=[]
rating=[]
year=[]

for i in soup.find_all('div',class_='ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 iALATN dli-title'):
    name.append(i.text)

for i in soup.find_all('a',class_="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating"):
    rating.append(i.text)
    
for i in soup.find_all('div',class_='sc-b189961a-7 feoqjK dli-title-metadata'):
    year.append(i.text)

df=pd.DataFrame({'Nmae':name,'Rating':rating,'Year of release':year})
df
```

**Question 2**- Write a python program to scrape details of all the posts from https://www.patreon.com/coreyms .Scrape the
heading, date, content and the likes for the video from the link for the youtube video from the post.

```python
#Answer 2
#Before executing this cell, execute the first cell to import the required libraries.
page=requests.get('https://www.patreon.com/coreyms')
soup=BeautifulSoup(page.content,'html.parser')

heading=[]
date=[]
content=[]
like=[]

for i in soup.find_all('div',class_="sc-bBHxTw iloeMK"):
    date.append(i.text)

for i in soup.find_all('a',class_="sc-kfPuZi doifNG"):
    date.append(i.text)
    
for i in soup.find_all('div',class_='sc-cfnzm4-0 daxSFj'):
    content.append(i.text)
    
for i in soup.find_all('div',class_='sc-jrQzAO knZGTZ'):
    like.append(i.text)

df=pd.DataFrame({'Heading':heading,'Date':date,'Content':content,'Like':like})
df
```

**Question 4**- Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/bestseller?sort=popular .

```python
#Answer 4
#Before executing this cell, execute the first cell to import the required libraries.
page=requests.get('https://www.bewakoof.com/bestseller?sort=popular')
soup=BeautifulSoup(page.content)

Name=[]
price=[]
imageURL=[]

for i in soup.find_all('div',class_='productNaming bkf-ellipsis' ,limit=10):
    Name.append(i.text)

for i in soup.find_all('div',class_="discountedPriceText" ,limit=10):
    price.append(i.text)
    
for i in soup.find_all('img',class_='productImgTag' ,limit=10):
    imageURL.append(i.get('src'))
df=pd.DataFrame({'Name':Name,'Price':price, 'Image URL':imageURL})
df
```

**Question 5**- Please visit https://www.cnbc.com/world/?region=world and scrap -
a) headings
b) date
c) News link

```python
#Answer 5
#Before executing this cell, execute the first cell to import the required libraries.
url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)
soup = BeautifulSoup(response.content)
heading=[]
date=[]
newslink=[]
for i in soup.find_all('div',class_='RiverHeadline-headline RiverHeadline-hasThumbnail'):
    heading.append(i.text)
    
    if i.find('a')['href']=='/pro/':
        value=i.find_all('a')
        newslink.append(value[1]['href'])
    else:
        newslink.append(i.find('a')['href'])
    date.append('N/A') #not all articles Doesn't contains date

df=pd.DataFrame({'Heading':heading,'Link':newslink,'Date':date}) #'date':date
df
```

**Question 6**- Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloadedarticles/ and scrap - a) Paper title
 b) date
c) Author

```python
#Answer 6
#Before executing this cell, execute the first cell to import the required libraries.

url = "https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/"
response = requests.get(url)
soup = BeautifulSoup(response.content)
title=[]
date=[]
author=[]
for i in soup.find_all('h2',class_='article-title'):
    title.append(i.text[3:])
for i in soup.find_all('p',class_='article-date'):
    date.append(i.text)
for i in soup.find_all('p',class_='article-authors'):
    author.append(i.text)
df=pd.DataFrame({'Title':title,'Date':date,'Author':author})
df
```
