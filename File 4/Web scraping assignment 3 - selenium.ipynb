Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and 
salary filter. 
You have to scrape data for “Data Scientist” designation for first 10 job results. 
You have to scrape the job-title, job-location, company name, experience required. 
The location filter to be used is “Delhi/NCR”. The salary filter to be used is “3-6” lakhs 
The task will be done as shown in the below steps: 
1. first get the web page https://www.naukri.com/
2. Enter “Data Scientist” in “Skill, Designations, and Companies” field. 
3. Then click the search button. 
4. Then apply the location filter and salary filter by checking the respective boxes 
5. Then scrape the data for the first 10 jobs results you get. 
6. Finally create a dataframe of the scraped data. 

```python
#Answer1

driver=webdriver.Chrome()
driver.get("https://www.naukri.com/")
search=driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[1]/div/div/div/div[1]/div/input")
search.send_keys("Data Scientist")
search_btn=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search_btn.click()
time.sleep(2)
location=driver.find_element(By.XPATH,"/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[4]/div[2]/div[3]")
location.click()
time.sleep(2)
salary=driver.find_element(By.XPATH,"/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[5]/div[2]/div[2]")
salary.click()

time.sleep(2)

JobTitle=[]
JobLocation=[]
CompanyName=[]
ExperienceRequired=[]

jobtitle=driver.find_elements(By.XPATH,"//div[contains(@class,'row1')]/a")
joblocation=driver.find_elements(By.XPATH,"//span[contains(@class,'loc-wrap ver-line')]")
companyname=driver.find_elements(By.XPATH,"//a[contains(@class,'comp-name')]")
experience=driver.find_elements(By.XPATH,"//span[contains(@class,'expwdth')]")

for i in jobtitle[:10]:
    JobTitle.append(i.text)
for j in joblocation[:10]:
    JobLocation.append(j.text)
for k in companyname[:10]:
    CompanyName.append(k.text)
for n in experience[:10]:
    ExperienceRequired.append(n.text)

driver.close()
    
data=pd.DataFrame({"Job Title":JobTitle,"Job Location":JobLocation,"Company Name":CompanyName,"Experience Required":ExperienceRequired})
data
```

Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuote
3. Than scrap a)Quote b) Author c) Type Of Quotes


```python
#Answer 6

driver=webdriver.Chrome()
driver.get("https://www.azquotes.com")
topquote=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
topquote.click()
time.sleep(2)
Quotes=[]
Author=[]
TypeOfQuote=[]
for n in range(10):
    if(n<9):
        page=driver.find_element(By.XPATH,"//li[contains(@class,'next')]/a[1]")
    quote=driver.find_elements(By.XPATH,"//div[contains(@class,'wrap-block')]/p[1]/a[2]")
    author=driver.find_elements(By.XPATH,"//div[contains(@class,'wrap-block')]/div[1]/a[1]")
    typeofquote=driver.find_elements(By.XPATH,"//div[contains(@class,'wrap-block')]/div[2]/div[1]")
    for i in quote:
        Quotes.append(i.text)
    for j in author:
        Author.append(j.text)
    for k in typeofquote:
        TypeOfQuote.append(k.text)
    if(n<9):
        page.click()

driver.close()
Data=pd.DataFrame({"Quote":Quotes,"Author":Author,"Type of Quote":TypeOfQuote})
Data
```

Q7: Write a python program to display list of respected former Prime Ministers of India (i.e. Name,
Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1
scrap the mentioned data and make the DataFrame

```python
#Answer 7

driver=webdriver.Chrome()
driver.get("https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1")
Name=[]
BornDead=[]
TermOfOffice=[]
Remarks=[]


for i in range(2,21):
    path1=str("//div[contains(@class,'TableData')]/table/tbody/tr["+str(i)+"]/td[2]/div")
    path2=str("//div[contains(@class,'TableData')]/table/tbody/tr["+str(i)+"]/td[3]/div")
    path3=str("//div[contains(@class,'TableData')]/table/tbody/tr["+str(i)+"]/td[4]")
    path4=str("//div[contains(@class,'TableData')]/table/tbody/tr["+str(i)+"]/td[5]/div")
    
    name=driver.find_element(By.XPATH,path1)
    Name.append(name.text)
    
    borndead=driver.find_element(By.XPATH,path2)
    BornDead.append(borndead.text)
    
    termofoffice=driver.find_element(By.XPATH,path3)
    TermOfOffice.append(termofoffice.text)
    
    remark=driver.find_element(By.XPATH,path4)
    Remarks.append(remark.text)

driver.close()
    
data=pd.DataFrame({"Name":Name,"Born-Dead":BornDead,"Term Of Office":TermOfOffice,"Remarks":Remarks})
data
```
