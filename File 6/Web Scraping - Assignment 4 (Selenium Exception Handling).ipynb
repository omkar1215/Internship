1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url 
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank 
B) Name 
C) Artist 
D) Upload date 
E) Views 

```python
driver=webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
time.sleep(2)

Rank =[]
Name =[]
Artist =[]
Upload_date=[]
Views=[]

try:
    name=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
    for i in name:
        Name.append(i.text)
    for j in range(len(name)):
        Rank.append(j+1)
except NoSuchElementExceptionhElementException:
    Name.append('-')
except StaleElementReferenceException:
    Name.append('-')

    
try:
    artist=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in artist:
        Artist.append(i.text)
except NoSuchElementExceptionhElementException:
    Artist.append('-')
except StaleElementReferenceException:
    Artist.append('-')

try:
    upload_date=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in upload_date:
        Upload_date.append(i.text)
except NoSuchElementExceptionhElementException:
    Upload_date.append('-')
except StaleElementReferenceException:
    Upload_date.append('-') 

try:
    views=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in views:
        Views.append(i.text)
except NoSuchElementExceptionhElementException:
    Views.append('-')
except StaleElementReferenceException:
    Views.append('-')
    
driver.close()
Data=pd.DataFrame({'Rank':Rank,'Video Name':Name,'Artist':Artist,'Views':Views,'Upload Date':Upload_date})
Data
```

5. Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/ You have to find the 
following details: 
A) Song name 
B) Artist name 
C) Last week rank 
D) Peak rank 
E) Weeks on board 

```python
driver=webdriver.Chrome()
driver.get('https://www.billboard.com/')
time.sleep(2)

Song =[]
Artist =[]
Last_Week_rank=[]
Peak_rank =[]
Weeks =[]

hot_100=driver.find_element(By.XPATH,'/html/body/div[3]/header/div/div[3]/div/nav/ul/li[1]/a')
try:
    hot_100.click()
except ElementNotInteractableException:
    driver.get(hot_100.get_attribute('href'))
time.sleep(2)

name=driver.find_element(By.XPATH,"//h3[contains(@class,'c-title  a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 u-font-size-23@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-245 u-max-width-230@tablet-only u-letter-spacing-0028@tablet')]")
Song.append(name.text)
try:
    song=driver.find_elements(By.XPATH,"//h3[contains(@class,'c-title  a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 lrv-u-font-size-18@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-330 u-max-width-230@tablet-only')]")
    for i in song:
        Song.append(i.text)
except NoSuchElementException:
    Song.append('-')
    
try:
    artist=driver.find_elements(By.XPATH,"//span[contains(@class,'c-label  a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only')]")
    for j in artist:
        Artist.append(j.text)
except NoSuchElementException:
    Artist.append('-')
    

try:
    last_rank=driver.find_elements(By.XPATH,"//li[contains(@class,'o-chart-results-list__item // a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-background-color-white-064@mobile-max u-hidden@mobile-max')]/span[1]")
    m=0
    for k in last_rank:
        if(m%2)==0:
            m+=1
            Last_Week_rank.append(k.text)
        else:
            m+=1
            Weeks.append(k.text)
except NoSuchElementException:
    Last_Week_rank.append('-')
    Weeks.append('-')

try:
    peak=driver.find_elements(By.XPATH,"//li[contains(@class,'o-chart-results-list__item // a-chart-bg-color a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-background-color-grey-lightest lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-hidden@mobile-max')]/span")
    for l in peak:
        Peak_rank.append(l.text)
except NoSuchElementException:
    Peak_rank.append('-')

driver.close()
Billboard=pd.DataFrame({'Song Name':Song,'Artist Name':Artist,'Last week rank':Last_Week_rank,'Peak':Peak_rank,'Weeks on chart':Weeks})
Billboard
```

7. Scrape the details most watched tv series of all time from imdb.com. 
Url = https://www.imdb.com/list/ls095964455/ You have 
to find the following details: 
A) Name 
B) Year span 
C) Genre 
D) Run time 
E) Ratings 
F) Votes

```python
driver=webdriver.Chrome()
driver.get("https://www.imdb.com/list/ls512407256/")
time.sleep(2)

Name =[]
Year_Span=[]
Genre =[]
Run_Time =[]
Ratings =[]
Votes =[]


for i in range(100):
    driver.execute_script('window.scrollBy(0, 300)')
    time.sleep(1.5)
    btn=driver.find_elements(By.XPATH,"//button[contains(@class,'ipc-icon-button dli-info-icon ipc-icon-button--base ipc-icon-button--onAccent2')]")
    btn[i].click()
    time.sleep(2)
    length=driver.find_element(By.XPATH,"/html/body/div[4]/div[2]/div/div[2]/div/div/div[1]/div[2]/ul[1]/li[2]")
    Run_Time.append(length.text)
    genre=driver.find_element(By.XPATH,"/html/body/div[4]/div[2]/div/div[2]/div/div/div[1]/div[2]/ul[2]")
    textt1=genre.text
    result=re.sub(r'([a-z])([A-Z])', r'\1 \2', textt1)
    Genre.append(result)
    closee=driver.find_element(By.XPATH,"//button[contains(@title,'Close Prompt')]")
    closee.click()
    
topclick=driver.find_element(By.XPATH,"/html/body/div[2]/button")
topclick.click()

for n in range(50):
    driver.execute_script('window.scrollBy(0, 500)')
    time.sleep(1.5)

try:
    name=driver.find_elements(By.XPATH,"//h3[contains(@class,'ipc-title__text')]")
    for i in name[:100]:
        textt1=i.text
        result1=textt1.split(' ', 1)
        Name.append(result1[1])
except NoSuchElementException:
    Name.append('-')

try:
    year=driver.find_elements(By.XPATH,"//div[contains(@class,'sc-b189961a-7 feoqjK dli-title-metadata')]/span[1]")
    for i in year:
        Year_Span.append(i.text)
except NoSuchElementException:
    Year_Span.append('-')
    

try:
    ratings=driver.find_elements(By.XPATH,"//span[contains(@class,'ipc-rating-star--rating')]")
    for i in ratings:
        Ratings.append(i.text)
except NoSuchElementException:
    Ratings.append('-')

try:
    votes=driver.find_elements(By.XPATH,"//span[contains(@class,'ipc-rating-star--voteCount')]")
    for i in votes:
        textt=i.text
        result=textt.replace("(", "").replace(")", "")
        Votes.append(result)
except NoSuchElementException:
    Votes.append('-')

driver.close()
Data=pd.DataFrame({"Name":Name[:100],"Year Span":Year_Span[:100],"Genre":Genre[:100],"Run Time":Run_Time[:100],"Ratings":Ratings[:100],"Votes":Votes[:100]})
Data
```

8. Details of Datasets from UCI machine learning repositories. 
Url = https://archive.ics.uci.edu/ You 
have to find the following details: 
A) Dataset name 
B) Data type 
C) Task 
D) Attribute type 
E) No of instances 
F) No of attribute G) Year 

```python
driver=webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
time.sleep(2)

dataset=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]")
dataset.click()
time.sleep(4)
accept=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/div/div[2]/button")
accept.click()
expand=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]")
expand.click()
time.sleep(2)

Dataset =[]
Data_Type =[]
Task =[]
Attribute_Type =[]
No_of_Instances =[]
No_of_Attribute =[]
Year =[]
for z in range(67):
    try:
        dataset=driver.find_elements(By.XPATH,"//h2[contains(@class,'truncate text-primary')]/a")
        for i in dataset:
            Dataset.append(i.text)
    except NoSuchElementException:
        Dataset.append('-')
    
    try:
        Type=driver.find_elements(By.XPATH,"//div[contains(@class,'my-2 hidden gap-4 md:grid grid-cols-12')]/div[2]/span")
        for j in Type:
            Data_Type.append(j.text)
    except NoSuchElementException:
        Data_Type.append('-')
    
    try:
        task=driver.find_elements(By.XPATH,"//div[contains(@class,'my-2 hidden gap-4 md:grid grid-cols-12')]/div[1]/span")
        for k in task:
            Task.append(k.text)
    except NoSuchElementException:
        Task.append('-')
    
    try:
        attribute_Type=driver.find_elements(By.XPATH,"//tbody[contains(@class,'border')]/tr/td[2]")
        for l in attribute_Type:
            if(l!=None):
                Attribute_Type.append(l.text)
            else:
                Attribute_Type.append('-')
    except NoSuchElementException:
        Attribute_Type.append('-')
    
    try:
        no_of_Instances=driver.find_elements(By.XPATH,"//div[contains(@class,'my-2 hidden gap-4 md:grid grid-cols-12')]/div[3]/span")
        for m in no_of_Instances:
            textt1=m.text
            result1= textt1.split(' ', 1)
            No_of_Instances.append(result1[0])
    except NoSuchElementException:
        No_of_Instances.append('-')

    try:
        no_of_Attribute=driver.find_elements(By.XPATH,"//div[contains(@class,'my-2 hidden gap-4 md:grid grid-cols-12')]/div[4]/span")
        for n in no_of_Attribute:
            text2=n.text
            result2=text2.split(' ', 1)
            No_of_Attribute.append(result2[0])
    except NoSuchElementException:
        No_of_Attribute.append('-')
    
    try:
        year=driver.find_elements(By.XPATH,"//tbody[contains(@class,'border')]/tr/td[3]")
        for b in year:
            if(b!=None):
                text3=b.text
                result3=text3.split("/")
                if(len(result3)==3):
                    Year.append(result3[2])
                else:
                    Year.append('-')
            else:
                Year.append('-')
    except NoSuchElementException:
        Year.append('-')
    if(z<66):
        nextt=driver.find_element(By.XPATH,"//div[contains(@class,'btn-group')]/button[2]")
        nextt.click()
        time.sleep(2)
        expand=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]")
        expand.click()
        expand=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]")
        expand.click()
        time.sleep(2)
    
        
driver.close()
UCI_Dataset=pd.DataFrame({'DataSet Title':Dataset[:600],'Data Type':Data_Type[:600],'Task':Task[:600],'Attribute Type':Attribute_Type[:600],'No of Instances':No_of_Instances[:600],'No of Attribute':No_of_Attribute[:600],'Year':Year[:600]})
UCI_Dataset
```
